```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import math

# import csv
python_linux_data = pd.read_csv('python_linux.csv')
matlab_linux_data = pd.read_csv('matlab_linux.csv')
octave_linux_data = pd.read_csv('octave_linux.csv')
python_win_data = pd.read_csv('python_win.csv')
matlab_win_data = pd.read_csv('matlab_win.csv')
octave_win_data = pd.read_csv('octave_win.csv')

# order by NNZ values
python_linux_data.sort_values(by=['NNZ'])
matlab_linux_data.sort_values(by=['NNZ'])
octave_linux_data.sort_values(by=['NNZ'])
python_win_data.sort_values(by=['NNZ'])
matlab_win_data.sort_values(by=['NNZ'])
octave_win_data.sort_values(by=['NNZ'])

# get matrix names for x axis on plots
x_py_linux = python_linux_data.Matrix
x_mat_linux = matlab_linux_data.Matrix
x_octave_linux = octave_linux_data.Matrix
x_py_win = python_win_data.Matrix
x_mat_win = matlab_win_data.Matrix
x_octave_win = octave_win_data.Matrix

#colors
color_py_linux = '#00b894'
color_mat_linux = '#0984e3'
color_octave_linux = '#fdcb6e'
color_py_windows = '#55efc4'
color_mat_windows = '#74b9ff'
color_octave_windows = '#ffeaa7'
```

```python
y_py_linux = python_linux_data.Memory
y_mat_linux = matlab_linux_data.Memory
y_octave_linux = octave_linux_data.Memory
y_py_win = python_win_data.Memory
y_mat_win = matlab_win_data.Memory
y_octave_win = octave_win_data.Memory


fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", y=y_py_linux, x=x_py_linux, name="Python Linux", marker_color=color_py_linux))
fig.add_trace(go.Histogram(histfunc="sum",y=y_py_win, x=x_py_win, name="Python Windows", marker_color=color_py_windows ))
fig.add_trace(go.Histogram(histfunc="sum",y=y_octave_linux, x=x_octave_linux, name="Octave Linux", marker_color=color_octave_linux))
fig.add_trace(go.Histogram(histfunc="sum", y=y_octave_win, x=x_octave_win, name="Octave Windows", marker_color=color_octave_windows))
fig.add_trace(go.Histogram(histfunc="sum",y=y_mat_linux, x=x_mat_linux, name="Matlab Linux", marker_color=color_mat_linux))
fig.add_trace(go.Histogram(histfunc="sum", y=y_mat_win, x=x_mat_win, name="Matlab Windows", marker_color=color_mat_windows))


#fig.update_yaxes(type="log")
fig.update_yaxes(title_text='Memory (MB)')
fig.update_layout(
    title={
        'text': "Memory comparison",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.show()
```

```python
y_py_linux = python_linux_data.RelError
y_mat_linux = matlab_linux_data.RelError
y_octave_linux = octave_linux_data.RelError
y_py_win = python_win_data.RelError
y_mat_win = matlab_win_data.RelError
y_octave_win = octave_win_data.RelError

fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", y=y_py_linux, x=x_py_linux, name="Python Linux", marker_color=color_py_linux))
fig.add_trace(go.Histogram(histfunc="sum",y=y_py_win, x=x_py_win, name="Python Windows", marker_color=color_py_windows ))
fig.add_trace(go.Histogram(histfunc="sum",y=y_octave_linux, x=x_octave_linux, name="Octave Linux", marker_color=color_octave_linux))
fig.add_trace(go.Histogram(histfunc="sum", y=y_octave_win, x=x_octave_win, name="Octave Windows", marker_color=color_octave_windows))
fig.add_trace(go.Histogram(histfunc="sum",y=y_mat_linux, x=x_mat_linux, name="Matlab Linux", marker_color=color_mat_linux))
fig.add_trace(go.Histogram(histfunc="sum", y=y_mat_win, x=x_mat_win, name="Matlab Windows", marker_color=color_mat_windows))

fig.update_yaxes(type="log")
fig.update_yaxes(tickformat='e')
fig.update_yaxes(title_text='Relative error')
fig.update_layout(
    title={
        'text': "Relative error comparison",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.show()
```

```python
y_py_linux = python_linux_data.Time
y_mat_linux = matlab_linux_data.Time
y_octave_linux = octave_linux_data.Time
y_py_win = python_win_data.Time
y_mat_win = matlab_win_data.Time
y_octave_win = octave_win_data.Time

fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", y=y_py_linux, x=x_py_linux, name="Python Linux", marker_color=color_py_linux))
fig.add_trace(go.Histogram(histfunc="sum",y=y_py_win, x=x_py_win, name="Python Windows", marker_color=color_py_windows ))
fig.add_trace(go.Histogram(histfunc="sum",y=y_octave_linux, x=x_octave_linux, name="Octave Linux", marker_color=color_octave_linux))
fig.add_trace(go.Histogram(histfunc="sum", y=y_octave_win, x=x_octave_win, name="Octave Windows", marker_color=color_octave_windows))
fig.add_trace(go.Histogram(histfunc="sum",y=y_mat_linux, x=x_mat_linux, name="Matlab Linux", marker_color=color_mat_linux))
fig.add_trace(go.Histogram(histfunc="sum", y=y_mat_win, x=x_mat_win, name="Matlab Windows", marker_color=color_mat_windows))

fig.update_yaxes(type="log")
fig.update_yaxes(title_text='Time (s)')
fig.update_layout(
    title={
        'text': "Time comparison",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.show()
```
