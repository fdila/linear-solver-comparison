```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import math

python_linux_data = pd.read_csv('python_linux.csv')
matlab_linux_data = pd.read_csv('matlab_linux.csv')
octave_linux_data = pd.read_csv('octave_linux.csv')
#TODO windows

x_py_l = python_linux_data.Matrix
x_mat_l = matlab_linux_data.Matrix
x_octave_l = octave_linux_data.Matrix
```

```python
y_py_l = python_linux_data.Memory
y_mat_l = matlab_linux_data.Memory
y_octave_l = octave_linux_data.Memory

fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", y=y_py_l, x=x_py_l, name="Python Linux"))
fig.add_trace(go.Histogram(histfunc="sum",y=y_mat_l, x=x_mat_l, name="Matlab Linux"))
fig.add_trace(go.Histogram(histfunc="sum",y=y_octave_l, x=x_octave_l, name="Octave Linux"))
fig.update_yaxes(type="log")
fig.update_yaxes(title_text='Memory (MB)')
fig.show()
```

```python
y_py_l = python_linux_data.RelError
y_mat_l = matlab_linux_data.RelError
y_octave_l = octave_linux_data.RelError

fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", y=y_py_l, x=x_py_l, name="Python Linux"))
fig.add_trace(go.Histogram(histfunc="sum",y=y_mat_l, x=x_mat_l, name="Matlab Linux"))
fig.add_trace(go.Histogram(histfunc="sum",y=y_octave_l, x=x_octave_l, name="Octave Linux"))
fig.update_yaxes(type="log")
fig.update_yaxes(tickformat='e')
fig.update_yaxes(title_text='Relative error')
fig.show()
```

```python
y_py_l = python_linux_data.Time
y_mat_l = matlab_linux_data.Time
y_octave_l = octave_linux_data.Time

fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", y=y_py_l, x=x_py_l, name="Python Linux"))
fig.add_trace(go.Histogram(histfunc="sum",y=y_mat_l, x=x_mat_l, name="Matlab Linux"))
fig.add_trace(go.Histogram(histfunc="sum",y=y_octave_l, x=x_octave_l, name="Octave Linux"))
fig.update_yaxes(type="log")
fig.update_yaxes(title_text='Time (s)')
fig.show()
```
